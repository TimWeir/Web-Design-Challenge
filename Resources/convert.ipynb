```python
import pandas as pd
```

```python
ingest_file = "cities.csv"
```

```python
cities_df = pd.read_csv(ingest_file, encoding="ISO-8859-1")
cities_df = cities_df.drop(['City_ID','Date','Lat', 'Lng'], axis=1)
cities_df = cities_df[['City','Country','Cloudiness','Humidity','Max Temp','Wind Speed']]
cities_df
```

```python
cities_html = cities_df.to_html(bold_rows=True, header=True, index=False, index_names=False)
```

```python
html_file = open("cities.html", "w")
html_file.write(cities_html)
html_file.close()
```
